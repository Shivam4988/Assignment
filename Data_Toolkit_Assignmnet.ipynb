### Question 1: Demonstrate three different methods for creating identical 2D arrays in NumPy. Provide the code for each method and the final output after each method.


Here are three different methods to create identical 2D arrays in NumPy:

```python
import numpy as np

# Method 1: Using np.array()
array1 = np.array([[1, 2, 3], [4, 5, 6]])
print("Method 1:\n", array1)

# Method 2: Using np.full()
array2 = np.full((2, 3), 7)
print("\nMethod 2:\n", array2)

# Method 3: Using np.ones() multiplied by a constant
array3 = np.ones((2, 3)) * 7
print("\nMethod 3:\n", array3)
```

**Shallow Copy:** Copies references, so changes in one affect the other.

**Deep Copy:** Creates a separate object, so changes don’t affect the original.

### Question 5: Create a NumPy array with random integers between 1 and 10 of shape (5, 6). After creating the array, perform the following operations:

(a) Extract all even integers from the array.

(b) Extract all odd integers from the array.

```python
arr = np.random.randint(1, 11, (5, 6))

# Extract even numbers
evens = arr[arr % 2 == 0]

# Extract odd numbers
odds = arr[arr % 2 != 0]

print("Array:\n", arr)
print("Even numbers:\n", evens)
print("Odd numbers:\n", odds)
```

### Question 7: Clean and transform the 'Phone' column in the sample dataset to remove non-numeric characters and convert it to a numeric data type. Also, display the table attributes and data types of each column.

```python
import pandas as pd


file_path = "/content/People Data.csv"

df_original = pd.read_csv(file_path)

# Remove non-numeric characters from the 'Phone' column and convert it to numeric
df_original["Phone"] = df_original["Phone"].astype(str).str.replace(r"\D", "", regex=True)

# Convert 'Phone' column to numeric (float) for consistency
df_original["Phone"] = pd.to_numeric(df_original["Phone"], errors='coerce')

# Display data types of each column
print(df_original.dtypes)

```

### Question 13: Create two NumPy arrays, x and y, each containing 100 random float values between 0 and 1. Perform the following tasks using Matplotlib and NumPy:

(a) Create a scatter plot using x and y, setting the color of the points to red and the marker style to 'o'.

(b) Add a horizontal line at y = 0.5 using a dashed line style and label it as 'y = 0.5'.

(c) Add a vertical line at x = 0.5 using a dotted line style and label it as 'x = 0.5'.

(d) Label the x-axis as 'X-axis' and the y-axis as 'Y-axis'.

(e) Set the title of the plot as 'Advanced Scatter Plot of Random Values'.

```python
import matplotlib.pyplot as plt

# Generate random data
x = np.random.rand(100)
y = np.random.rand(100)

# (a) Create scatter plot
plt.scatter(x, y, color='red', marker='o', label='Random Points')

# (b) Add a horizontal line at y=0.5
plt.axhline(y=0.5, linestyle='--', color='blue', label='y = 0.5')

# (c) Add a vertical line at x=0.5
plt.axvline(x=0.5, linestyle=':', color='green', label='x = 0.5')

# (d) Label the axes
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# (e) Set the title
plt.title("Advanced Scatter Plot of Random Values")

# (f) Display the legend
plt.legend()

# Show the plot
plt.show()
```

### Question 14: Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity', and perform the following tasks using Matplotlib:



(a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and right y-axis for 'Humidity').

(b) Label the x-axis as 'Date'.

(c) Set the title of the plot as 'Temperature and Humidity Over Time'.

```python
# Generate time-series data
dates = pd.date_range(start="2024-01-01", periods=100)
df_time = pd.DataFrame({
    "Date": dates,
    "Temperature": np.random.randint(20, 35, 100),
    "Humidity": np.random.randint(30, 90, 100)
})

# Create a dual-axis plot
fig, ax1 = plt.subplots()

# (a) Plot Temperature on left y-axis
ax1.plot(df_time["Date"], df_time["Temperature"], 'r-', label="Temperature")
ax1.set_ylabel("Temperature")
ax1.set_xlabel("Date")

# Create second y-axis for Humidity
ax2 = ax1.twinx()
ax2.plot(df_time["Date"], df_time["Humidity"], 'b-', label="Humidity")
ax2.set_ylabel("Humidity")

# (c) Set the title
plt.title("Temperature and Humidity Over Time")

# Show the plot
plt.show()
```

### Question 16: Using Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()  # Display in Jupyter Notebook

# Generate data for sine wave
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

# Create Bokeh figure
p = figure(title="Sine Wave Function", x_axis_label="x", y_axis_label="sin(x)")
p.line(x, y, line_width=2)

# Show plot
show(p)
```

### Question 17: Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart'.


```python
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.io import output_notebook

output_notebook()

# Generate sample categorical data
categories = ["A", "B", "C", "D", "E"]
values = np.random.randint(10, 100, size=5)

# Create data source
source = ColumnDataSource(data=dict(categories=categories, values=values))

# Create figure
p = figure(x_range=categories, title="Random Categorical Bar Chart",
           x_axis_label="Category", y_axis_label="Value")

# Add bars with hover tool
p.vbar(x='categories', top='values', width=0.5, source=source, color="navy")

hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

# Show plot
show(p)
```
